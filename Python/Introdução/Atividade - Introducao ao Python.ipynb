```python
"""
1. Faça um programa que tenha uma função chamada amplitude. A função deve receber uma lista e imprimir a amplitude.
Crie também um código para testar sua função
"""
import numpy as np

x = [5, 9, 10, 14, 19, 24]

def amplitude():
    min = np.min(x)
    max = np.max(x)
    print(max - min)
amplitude()
```

```python
"""
2. Faça uma função que receba uma string e imprima esta string na forma vertical
Por exemplo, se receber python, deve imprimir
p
y
t
h
o
n
Dica: uma string do python funciona como uma lista!
Crie também um código para testar sua função
"""

palavra = input("Digite uma palavra: ")

def imprimir(texto):
    for n in range(0, len(texto)):
        print(texto[n])
imprimir(palavra)
```

```python
"""
3. Crie um programa que leia o peso de uma carga em números inteiros. Se o peso for até 10 kg, informe que o valor será de R$ 50,00. Entre 11 e 20 kg, informe que o valor será de R$ 80. Se for maior
que  20 informe que o transporte não é aceito. Teste vários pesos.
"""

print("Informe o peso")
peso = int(input('Digite o Peso: '))

if(peso <= 10):
    print("O valor será de R$ 50,00")
elif(peso >= 11 and peso <= 20):
    print("O valor será de R$ 80,00")
else:
    print("O transporte não é aceito")
```
